### MENGUBAH TIPE DATA KOLOM

Berdasarkan informasi data tersebut, tipe dari data tersebut adalah object sehingga akan diganti dengan float untuk membantu mempermudah analisis dan pengolahan data.

```python
#mengubah tipe data object menjadi float untuk kolom Happiness score hingga Explained by: Perceptions of corruption
obj_col = data.select_dtypes(include = ['O']).columns

columns = [col for col in obj_col if col not in ['RANK', 'Country']]

for col in columns:
    data[col] = data[col].str.replace(',','.').astype(float)
```

```python
data_western_europe = data[data['Country'].isin(western_europe)]
data_northern_europe = data[data['Country'].isin(northern_europe)]
data_southern_europe = data[data['Country'].isin(southern_europe)]
data_eastern_europe =data[data['Country'].isin(eastern_europe)]
```

Akan dilihat apakah setiap wilayah di Benua Eropa memiliki happiness score yang tinggi dan lebih dari rata-rata.

```python
#melihat rata-rata nilai happiness score 
mean_happiness_score = data['Happiness score'].mean()
mean_happiness_score
```

```python
#Eropa Barat
#membuat figur dan sumbu untuk plot
fig, ax = plt.subplots(figsize=(8, 7))

#membuat grafik bar untuk menampilkan skor kebahagiaan di setiap negara di Eropa Barat
bars1 = ax.bar(data_western_europe['Country'], data_western_europe['Happiness score'], color = colors_orange, alpha=0.4, edgecolor='red')

#mengatur penamaan dalam diagram
plt.title('Faktor Happiness Score di Wilayah Eropa Barat')

#mengatur rotasi label sumbu x agar lebih mudah dibaca
ax.set_xticklabels(data_western_europe['Country'], rotation=45, ha='right')

#mengatur label sumbu x
ax.set_xlabel("Country", fontsize=14, labelpad=10, fontweight='bold', color= 'black')

#mengatur label sumbu y
ax.set_ylabel("Happiness Score", fontsize=14, labelpad=10, fontweight='bold', color= 'black')

#mendapatkan batas-batas sumbu x dan y
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

#menambahkan garis putus-putus untuk menandakan skor kebahagiaan rata-rata (Global Mean)
plt.axhline(mean_happiness_score, linestyle='--', color='black', label='Global Mean')

#menambahkan teks "Global Mean" di luar batas sumbu x
ax.text(xmax + 0.1, mean_happiness_score, f'Global Mean: {mean_happiness_score:.2f}', ha='left', va='center', fontsize=12, color='red')

#menampilkan nilai skor kebahagiaan di atas setiap bar
for index, value in enumerate(data_western_europe['Happiness score']):
    plt.text(index, value + 0.1, f'{value:.2f}', ha='center', va='bottom', fontsize=9)

#menampilkan legenda
plt.legend()

plt.show()
```

```python
#mengambil negara dengan happiness score tertinggi dari masing-masing wilayah
top_country_western_europe = data_western_europe.loc[data_western_europe['Happiness score'].idxmax()]
top_country_northern_europe = data_northern_europe.loc[data_northern_europe['Happiness score'].idxmax()]
top_country_southern_europe = data_southern_europe.loc[data_southern_europe['Happiness score'].idxmax()]
top_country_eastern_europe = data_eastern_europe.loc[data_eastern_europe['Happiness score'].idxmax()]

top_countries_europe_df = pd.DataFrame({
    'Region': ['Eropa Barat', 'Eropa Utara', 'Eropa Selatan', 'Eropa Timur'],
    'Country': [top_country_western_europe['Country'], top_country_northern_europe['Country'],
                top_country_southern_europe['Country'], top_country_eastern_europe['Country']],
    'Happiness Score': [top_country_western_europe['Happiness score'], top_country_northern_europe['Happiness score'],
                top_country_southern_europe['Happiness score'], top_country_eastern_europe['Happiness score']],
    'Rank': [top_country_western_europe['RANK'], top_country_northern_europe['RANK'],
             top_country_southern_europe['RANK'], top_country_eastern_europe['RANK']]
})


# Mengurutkan DataFrame berdasarkan Happiness Score secara menurun
top_countries_europe_df = top_countries_europe_df.sort_values(by='Happiness Score', ascending=False)

# Mengatur ukuran figur grafik
plt.figure(figsize=(12, 8))

# Membuat plot untuk menampilkan negara dengan happiness score tertinggi dari masing-masing wilayah
ax = sns.barplot(x='Region', y='Happiness Score', hue='Country', data=top_countries_europe_df, palette=colors_purple, dodge=False)

# Menambahkan nilai Happiness Score dan Rank pada setiap bar
for i, v in enumerate(top_countries_europe_df['Happiness Score']):
    plt.text(i, v + 0.25, f'{v:.2f}', ha='center', color='black', fontweight='bold')
    plt.text(i, v - 0.5, f'Rank: {top_countries_europe_df.iloc[i]["Rank"]}', ha='center', color='black', fontstyle='italic')

# Menetapkan batasan rentang sumbu y agar lebih luas
plt.ylim(0, top_countries_europe_df['Happiness Score'].max() + 0.75)    

# Mendapatkan batas-batas sumbu x dan y
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

# Tambahkan garis putus-putus untuk menandakan skor kebahagiaan rata-rata (Global Mean)
plt.axhline(mean_happiness_score, linestyle='--', color='red', label='Global Mean')

# Tambahkan teks "Global Mean" di luar batas sumbu x
ax.text(xmax + 0.1, mean_happiness_score, f'Global Mean: {mean_happiness_score:.2f}', ha='left', va='center', fontsize=12, color='red')

plt.title('Negara dengan Happiness Score Tertinggi dari Setiap Wilayah Eropa', fontsize=14)
plt.xlabel('Wilayah', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend(title= 'Negara', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()  # Mengatur layout agar rapi
plt.show()
```

```python
#mengambil negara dengan happiness score tertinggi dari masing-masing wilayah
top_country_western_europe = data_western_europe.loc[data_western_europe['Happiness score'].idxmax()]
top_country_northern_europe = data_northern_europe.loc[data_northern_europe['Happiness score'].idxmax()]
top_country_southern_europe = data_southern_europe.loc[data_southern_europe['Happiness score'].idxmax()]
top_country_eastern_europe = data_eastern_europe.loc[data_eastern_europe['Happiness score'].idxmax()]

#menggabungkan negara-negara dengan happiness score tertinggi dari setiap wilayah menjadi satu dataframe
data_plot_final = pd.concat([top_country_western_europe, top_country_northern_europe, top_country_southern_europe, top_country_eastern_europe])

#columns to include in the stacked bar chart
columns = ['Dystopia (1.83) + residual', 'Explained by: GDP per capita',
           'Explained by: Social support', 'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity', 'Explained by: Perceptions of corruption']

#set the labels for each category
labels = columns

#set the positions for the bars
x = np.arange(len(labels))

#set the width of the bars
width = 0.15

#plot the stacked bars
fig, ax = plt.subplots(figsize=(12, 8))

#set data for each country
category_western = top_country_western_europe[columns].values.flatten()
category_northern = top_country_northern_europe[columns].values.flatten()
category_southern = top_country_southern_europe[columns].values.flatten()
category_eastern = top_country_eastern_europe[columns].values.flatten()

#plot bars for each region
bars_west = ax.barh(x - 3*width/2, category_western, width, label='Western Europe', color=colors_purple[3])
bars_north = ax.barh(x - width/2, category_northern, width, label='Northern Europe', color=colors_purple[2])
bars_south = ax.barh(x + width/2, category_southern, width, label='Southern Europe', color=colors_purple[1])
bars_east = ax.barh(x + 3*width/2, category_eastern, width, label='Eastern Europe', color=colors_purple[0])

#add labels to the bars
for i, (category, value) in enumerate(zip(labels, category_western)):
    ax.text(value, i - 3*width/2, f'{value:.2f}', ha='left', va='center', color='black')

for i, (category, value) in enumerate(zip(labels, category_northern)):
    ax.text(value, i - width/2, f'{value:.2f}', ha='left', va='center', color='black')

for i, (category, value) in enumerate(zip(labels, category_southern)):
    ax.text(value, i + width/2, f'{value:.2f}', ha='left', va='center', color='black')

for i, (category, value) in enumerate(zip(labels, category_eastern)):
    ax.text(value, i + 3*width/2, f'{value:.2f}', ha='left', va='center', color='black')


#add labels, title, and legend
ax.set_xlabel('Nilai skor')
ax.set_ylabel('Kategori')
ax.set_title('Perbandingan Happiness Score di Top Country Benua Eropa')
ax.set_yticks(x)
ax.set_yticklabels(labels)
ax.legend()

#adjust layout
plt.tight_layout()
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

#show plot
plt.show()
```
